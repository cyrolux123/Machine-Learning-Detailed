```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('aff.csv', header = 1)
dataset.head()
```

```python
dataset.info()
```

Data Cleaning

```python
## missing values
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
```

```python
df.head()
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# Remove the row 122
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.columns
```

```python
## Change the required columns as integer data type
df[['month', 'day', 'year', 'RH', 'Ws']] = df[['month', 'day', 'year', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
## Changing the other columns as float data type
objects = [features for features in df.columns if df[features].dtype == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# Saving the cleaned dataset
df.to_csv('cleaned_aff.csv', index = False)
```

```python
df['Classes'] = np.where(df['Classes'].astype(str).str.contains('not fire'), 'not fire', 'fire')
```

```python
# Monthly Fire Analysis
dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(15, 5))
sns.set(style='whitegrid')
sns.countplot(x='month', data=df, hue='Classes', palette='viridis')
plt.title('Analysis of Bejaia Region')
plt.ylabel('Number of Fires')
plt.xlabel('Months')
plt.show()
```

```python
# Monthly Fire Analysis
dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(15, 5))
sns.set(style='whitegrid')
sns.countplot(x='month', data=df, hue='Classes', palette='viridis')
plt.title('Analysis of Brjaia Region')
plt.ylabel('Number of Fires')
plt.xlabel('Months')
plt.show()
```

It is observed that August and September had the most number of forest fires for both regions. And from the above plot of months, we can understand few things.

Most of the fires happened in August and very high Fires happened in only 3 months - June, July and August. Less Fires was on September.
