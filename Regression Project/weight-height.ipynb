Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the dataset
df = pd.read_csv('height-weight.csv')
df.head()
```

```python
plt.scatter(df['Weight'], df['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight vs Height')
plt.show()
```

```python
## Divide the dataset into independent and dependent variables
X = df[['Weight']] # Independent feature
y = df[['Height']] # Dependent feature
```

```python
X, y
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
X.shape
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
## Standardize the dataset over independent train data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
plt.scatter(X_train, y_train)
```

```python
# Train the model Simple Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print("The slope or coefficient of weight is: ", model.coef_)
print("The intercept of the line is: ", model.intercept_)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, model.predict(X_train), 'r')
```

```python
y_pred_test = model.predict(X_test)
```
